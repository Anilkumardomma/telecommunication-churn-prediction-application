# preprocessing

```python
# Helper function to avoid column name mismatches (dots vs underscores)
def real_col(df, name):
    candidates = [name, name.replace('.', ''), name.replace('.', ' '), name.replace('', '.')]
    for c in candidates:
        if c in df.columns:
            return c
    return None

#We added a helper function to handle column name mismatches in the dataset.
#It ensures that whether a column is written with dots, underscores, or spaces, we can still detect it correctly.
#This makes our preprocessing more robust
```

```python
# Drop redundant columns: charges and voicemail_messages
drops = [
    real_col(df, 'day_charge'),
    real_col(df, 'evening_charge'),
    real_col(df, 'night_charge'),
    real_col(df, 'international_charge'),
    real_col(df, 'total_charge'),
    real_col(df, 'voice_mail_messages')   # redundant with voice_mail_plan 
]
drops = [c for c in drops if c is not None]
print("Dropping:", drops)

df = df.drop(columns=drops)
print("Shape after drop:", df.shape) 

#We dropped 6 redundant columns: 5 charge variables (day, evening, night, international, and total charge) and voice_mail_messages.
#Charges are perfectly correlated with minutes, and voicemail messages are redundant with voicemail plan.
#After dropping, our dataset is reduced to 13 columns, removing multicollinearity and simplifying the model.
```

```python
# winsorization / capping to treat outliers
import numpy as np

num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
target = real_col(df, 'churn')
if target in num_cols:
    num_cols.remove(target)

print("Numeric columns to cap:", num_cols)

df_capped = df.copy()
low_q, high_q = 0.01, 0.99
for col in num_cols:
    low = df_capped[col].quantile(low_q)
    high = df_capped[col].quantile(high_q)
    df_capped[col] = df_capped[col].clip(lower=low, upper=high)

# quick sanity check
display(df[num_cols].describe().T[['min','max']].head())
display(df_capped[num_cols].describe().T[['min','max']].head())

#We applied Winsorization (1stâ€“99th percentile capping) to reduce the effect of outliers in numeric columns.
#For example, account_length was capped from 1â€“243 to 12â€“195, and day_mins from 0â€“350 to 52â€“305.
#This preserves most of the data distribution while minimizing the influence of extreme values.
```

```python
#Feature preparation
import numpy as np
import pandas as pd

target = real_col(df_capped, 'churn')
X = df_capped.drop(columns=[target])
y = df_capped[target].astype(int)

# Ensure X is numeric
X = X.select_dtypes(include=[np.number])

print("X shape:", X.shape, "y shape:", y.shape)

#We separated features X and target y. X contains 12 numeric predictors, while y contains the churn labels.
#This step ensures the dataset is in the correct format for model training
```

```python
#Trainâ€“Test Split
from sklearn.model_selection import train_test_split
RSEED = 42

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=RSEED
)

print("Train:", X_train.shape, "Test:", X_test.shape)
print("Train churn dist:\n", y_train.value_counts(normalize=True))

#We split the dataset into 75% training and 25% testing sets.
#Stratification was applied to maintain the original churn imbalance (85.5% non-churners, 14.5% churners) in both sets.
#This ensures fair evaluation of models
```

```python
##Handling Missing Values and Feature Scaling
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_features = X_train.columns.tolist()

imp = SimpleImputer(strategy='median')
scaler = StandardScaler()

# Train
X_train_num = imp.fit_transform(X_train[num_features])
X_train_scaled = scaler.fit_transform(X_train_num)

# Test
X_test_num = imp.transform(X_test[num_features])
X_test_scaled = scaler.transform(X_test_num)

# Back to DataFrames
X_train_proc = pd.DataFrame(X_train_scaled, columns=num_features, index=X_train.index)
X_test_proc  = pd.DataFrame(X_test_scaled, columns=num_features, index=X_test.index)

print("Preprocessing done:", X_train_proc.shape, X_test_proc.shape)


#We applied preprocessing to handle missing values and standardize features.
#Missing numeric values were imputed using the median, and all features were scaled to mean 0 and variance 1.
#This ensures consistency across features and improves performance for models sensitive to feature scales.
```

```python
#Handling class imbalance with SMOTE
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=RSEED)

X_train_res, y_train_res = sm.fit_resample(X_train_proc, y_train)

print("After SMOTE class counts:\n", y_train_res.value_counts())
print("Resampled X shape:", X_train_res.shape)


#We used SMOTE to handle class imbalance.
#Originally churners made up only 14.5% of the dataset. After SMOTE the training set is balanced with 2137 churners and 2137 non-churners.
#This prevents the model from being biased toward non-churners and improves recall for the minority churn class.
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Logistic Regression
lr = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=RSEED)
lr.fit(X_train_res, y_train_res)
y_pred_lr = lr.predict(X_test_proc)
y_prob_lr = lr.predict_proba(X_test_proc)[:,1]

print("Logistic Regression")
print(classification_report(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_lr))


#Logistic Regression achieved an accuracy of 76% and ROC-AUC of 0.81.
#It identified churners with recall of 71%, meaning it can catch most churners, but precision is low 34%,
#so it often misclassifies non-churners as churners.
#As a baseline model, Logistic Regression provides useful interpretability but is limited in predictive power compared to ensemble methods
```

```python
# Gradient Boosting
gb = GradientBoostingClassifier(random_state=RSEED)
gb.fit(X_train_res, y_train_res)

y_pred_gb = gb.predict(X_test_proc)
y_prob_gb = gb.predict_proba(X_test_proc)[:,1]

print("Gradient Boosting")
print(classification_report(y_test, y_pred_gb))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_gb))

#Gradient Boosting significantly improved results compared to Logistic Regression.
#Accuracy increased to 91% and ROC-AUC to 0.89. It identified 72% of churners with precision of 68%,
#a strong balance between catching churners and avoiding false alarms.
#This shows that tree-based ensemble methods are better suited for churn prediction than linear models
```

```python
# Random Forest
rf = RandomForestClassifier(n_estimators=200, random_state=RSEED, n_jobs=-1)
rf.fit(X_train_res, y_train_res)
y_pred_rf = rf.predict(X_test_proc)
y_prob_rf = rf.predict_proba(X_test_proc)[:,1]

print("\nRandom Forest")
print(classification_report(y_test, y_pred_rf))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))

#Random Forest achieved 92% accuracy and ROC-AUC of 0.89.
#It detected 73% of churners with precision of 71%, outperforming Logistic Regression and slightly improving over Gradient Boosting.
#This shows Random Forest is a strong model balancing accuracy and robustness.
```

```python
#Confusion matrix
models = {
    "Logistic Regression": y_pred_lr,
    "Gradient Boosting": y_pred_gb,
    "Random Forest": y_pred_rf
}

for name, y_pred in models.items():
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"Confusion Matrix ({name})")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

#Logistic Regression catches many churners (high recall) but generates lots of false alarms (low precision).
#Gradient Boosting reduces false alarms compared to Logistic Regression, while still detecting most churners.
#Random Forest performs slightly better than Gradient Boosting, missing fewer churners and achieving stronger balance overall
```

```python
from xgboost import XGBClassifier

# XGBoost
xgb = XGBClassifier(random_state=RSEED,eval_metric='logloss',use_label_encoder=False)

xgb.fit(X_train_res, y_train_res)

y_pred_xgb = xgb.predict(X_test_proc)
y_prob_xgb = xgb.predict_proba(X_test_proc)[:,1]

print("XGBoost")
print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_xgb))

#XGBoost outperformed all previous models
#achieving 94% accuracy and ROC-AUC of 0.90. It detected 78% of churners with precision of 79%,
#the best trade off between recall and precision so far. This makes XGBoost our final chosen model for churn prediction.
```

```python
# Confusion Matrix for XGBoost
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (XGBoost)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

#The XGBoost confusion matrix shows 688 true non-churns and 94 true churns correctly identified.
#Only 25 false positives and 27 false negatives occurred.
#This balance confirms that XGBoost is effective at catching churners while minimizing false alarms
```

```python
#Hyperparameter Tuning
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier

# Parameter grid (compact but effective)
param_dist = {
    'n_estimators': [100, 200, 300],       # number of trees
    'max_depth': [3, 4, 5, 6],             # depth of trees
    'learning_rate': [0.01, 0.05, 0.1, 0.2], # step size
    'subsample': [0.8, 1.0],               # % of rows used per tree
    'colsample_bytree': [0.8, 1.0]         # % of features used per tree
}

xgb_tuned = XGBClassifier(random_state=RSEED,eval_metric='logloss',use_label_encoder=False)

# Randomized search (tries 20 random combinations)
random_search = RandomizedSearchCV(estimator=xgb_tuned,param_distributions=param_dist,n_iter=20,scoring='roc_auc',cv=5,verbose=1,n_jobs=-1,random_state=RSEED)

# Fit tuning on balanced training data
random_search.fit(X_train_res, y_train_res)

# Best parameters and CV performance
print("Best parameters:", random_search.best_params_)
print("Best ROC-AUC (CV):", random_search.best_score_)


#Defined a parameter grid with values for number of trees (n_estimators), tree depth (max_depth), learning rate, and sampling ratios.
#Used RandomizedSearchCV with 5-fold cross-validation to test 20 random parameter combinations, optimizing for ROC-AUC
#It identified the best parameters (n_estimators=200, max_depth=6, learning_rate=0.1, subsample=0.8, colsample_bytree=0.8),
#Achieved a very strong cross-validation ROC-AUC of 0.994, showing that the tuned XGBoost is close to optimal.
```

```python
# Retrain best XGBoost model on full training data
best_xgb = random_search.best_estimator_

best_xgb.fit(X_train_res, y_train_res)

# Evaluate on test set
y_pred_best = best_xgb.predict(X_test_proc)
y_prob_best = best_xgb.predict_proba(X_test_proc)[:,1]

print("Tuned XGBoost Results")
print(classification_report(y_test, y_pred_best))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_best))

#Tuned XGBoost was retrained on the full balanced training data 
#using the best parameters from RandomizedSearchCV. 
#On the test set, it achieved 94% accuracy and ROC-AUC of 0.90. 
#It detected 75% of churners with precision of 80%, 
#confirming strong but slightly lower recall than baseline XGBoost.
```

```python
# Confusion Matrix for Tuned XGBoost
cm_best = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Tuned XGBoost)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

#The Tuned XGBoost confusion matrix shows 690 true non-churns 
#and 91 true churns correctly identified. 
#Only 23 false positives and 30 false negatives occurred. 
#This confirms that Tuned XGBoost is effective, 
#though it misses slightly more churners than baseline XGBoost.
```

```python
# Model Comparison
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score

# Collect predictions & probabilities
preds = {
    'Logistic Regression': (y_pred_lr, y_prob_lr),
    'Random Forest': (y_pred_rf, y_prob_rf),
    'Gradient Boosting': (y_pred_gb, y_prob_gb),
    'XGBoost': (y_pred_xgb, y_prob_xgb),
    'Tuned XGBoost': (y_pred_best, y_prob_best)   
}

# Build results table
rows = []
for name,(y_pred,y_prob) in preds.items():
    rows.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision (Churn=1)': precision_score(y_test, y_pred, pos_label=1),
        'Recall (Churn=1)': recall_score(y_test, y_pred, pos_label=1),
        'F1 (Churn=1)': f1_score(y_test, y_pred, pos_label=1),
        'ROC-AUC': roc_auc_score(y_test, y_prob)
    })

results_df = pd.DataFrame(rows).set_index('Model').round(3)
results_df


#Comparing all models Logistic Regression performed the least, with low precision (0.34) and F1 (0.46), making it unsuitable for churn prediction.
#Random Forest and Gradient Boosting gave solid results (approx 91â€“92% accuracy), but not the best among models.
#XGBoost achieved the highest accuracy (0.938) and the best F1 (0.783), showing strong balance between catching churners and avoiding false alarms.
#Tuned XGBoost performed almost the same, with slightly higher precision but lower recall, so no major improvement over baseline XGBoost.

#So XGBoost is selected as the best model for churn prediction, with 94% accuracy and strong overall performance.
```

```python
code= '''# Importing Libraries
# This block imports the required libraries for the application. These include Streamlit for the web framework,
# Pandas for data manipulation, Joblib for model loading, and Plotly and SHAP for visualization and model explainability.
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import plotly.graph_objects as go
import shap

# Page Configuration
# This code configures the web page's metadata and default layout. It sets the browser tab title,
# centers the main content for better readability, and ensures the sidebar is collapsed on initial load.
st.set_page_config(
    page_title="Churn Predictor",
    layout="centered",
    initial_sidebar_state="collapsed"
)

# Loading and Caching ML Assets
# This function loads the pre-trained machine learning assets required for prediction. The '@st.cache_resource'
# decorator is used to optimize performance by loading these assets into memory only once when the app starts.
@st.cache_resource
def load_assets():
    """Loads all necessary machine learning assets."""
    imputer = joblib.load("imputer.joblib")
    scaler = joblib.load("scaler.joblib")
    model = joblib.load("xgb_churn_model.joblib")
    explainer = shap.TreeExplainer(model)
    return imputer, scaler, model, explainer

imputer, scaler, model, explainer = load_assets()

# App Header
# This section renders the main title and a descriptive subtitle for the application. It provides the user
# with an immediate understanding of the app's purpose and scope.
st.title("Customer Churn Predictor")
st.write("Predict churn with high accuracy by providing the complete customer profile.")

# The Input Form
# This block creates a user input form with 'st.form' to efficiently gather customer data. Using a form ensures
# the prediction logic only runs upon submission, and widgets are organized into columns for a clean UI.
with st.form("churn_prediction_form"):
    st.header("Enter Full Customer Details")

    # Plans & Account
    st.subheader("Account and Plans")
    col1, col2 = st.columns(2)
    with col1:
        international_plan = st.radio("Has International Plan?", ("No", "Yes"), horizontal=True)
    with col2:
        voice_mail_plan = st.radio("Has Voice Mail Plan?", ("No", "Yes"), horizontal=True)
    
    account_length = st.slider("Account Length (days)", 1, 240, 100)
    customer_service_calls = st.slider("Customer Service Calls", 0, 10, 1)

    # Usage Details (Minutes)
    st.subheader("Usage Details (in Minutes)")
    col3, col4, col5 = st.columns(3)
    with col3:
        day_mins = st.number_input("Day Mins", min_value=0.0, value=180.0)
    with col4:
        evening_mins = st.number_input("Evening Mins", min_value=0.0, value=200.0)
    with col5:
        night_mins = st.number_input("Night Mins", min_value=0.0, value=200.0)
    
    international_mins = st.number_input("International Mins", min_value=0.0, value=10.0)
    
    # Call Details (Number of Calls)
    st.subheader("Number of Calls")
    col6, col7, col8 = st.columns(3)
    with col6:
        day_calls = st.number_input("Day Calls", min_value=0, value=100)
    with col7:
        evening_calls = st.number_input("Evening Calls", min_value=0, value=100)
    with col8:
        night_calls = st.number_input("Night Calls", min_value=0, value=100)

    international_calls = st.number_input("International Calls", min_value=0, value=4)

    # Submit Button
    st.markdown("---") # Visual separator
    predict_button = st.form_submit_button(label="Predict Churn", use_container_width=True)


# Prediction Logic
# Upon form submission, this logic preprocesses the user input data through the imputer and scaler.
# It then generates a churn probability using the trained model and calculates SHAP values to explain the prediction.
if predict_button:
    # Convert radio button inputs to numerical values
    int_plan_val = 1 if international_plan == "Yes" else 0
    vm_plan_val = 1 if voice_mail_plan == "Yes" else 0

    # Create the DataFrame with ALL user inputs 
    feature_names = [
        'account_length', 'voice_mail_plan', 'day_mins', 'evening_mins',
        'night_mins', 'international_mins', 'customer_service_calls',
        'international_plan', 'day_calls', 'evening_calls',
        'night_calls', 'international_calls'
    ]
    input_data = pd.DataFrame([[
        account_length, vm_plan_val, day_mins, evening_mins,
        night_mins, international_mins, customer_service_calls, int_plan_val,
        day_calls, evening_calls, night_calls, international_calls
    ]], columns=feature_names)

    # Preprocess and predict
    input_imputed = imputer.transform(input_data)
    input_scaled = scaler.transform(input_imputed)
    probability = model.predict_proba(input_scaled)[0][1]
    shap_values = explainer.shap_values(input_scaled)

    # Displaying the Prediction Result
    # The prediction results are presented in a two-column layout for clarity. It displays a color-coded
    # status (High/Low Risk) for quick interpretation and a precise 'st.metric' for the numerical churn probability.
    st.markdown("---")
    st.header("Prediction Result")
    
    res_col1, res_col2 = st.columns(2)
    with res_col1:
        if probability > 0.5:
            st.error("Status: HIGH RISK")
        else:
            st.success("Status: LOW RISK")
    with res_col2:
        st.metric(label="Churn Probability", value=f"{probability:.1%}")

    # Creating the Explanation Plot
    # This section enhances model transparency by visualizing the prediction's rationale with a SHAP bar chart. It shows
    # the top five features influencing the outcome, color-coded to indicate if they increased or decreased churn risk.
    st.subheader("Top 5 Reasons for this Prediction")
    
    # Creates a DataFrame for SHAP values and feature names.
    shap_df = pd.DataFrame({
        'feature': [f.replace('_', ' ').title() for f in input_data.columns],
        'shap_value': shap_values[0, :],
    })
    
    # Assign colors like red for features increasing churn risk, blue for those decreasing it.
    shap_df['color'] = np.where(shap_df['shap_value'] > 0, '#e53e3e', '#3182ce')
    
    # Sort by the absolute impact to find the top 5 features.
    shap_df['abs_shap'] = np.abs(shap_df['shap_value'])
    shap_df = shap_df.sort_values('abs_shap', ascending=False).head(5)

    # Creates and display the Plotly bar chart.
    fig = go.Figure(go.Bar(
        x=shap_df['shap_value'],
        y=shap_df['feature'],
        orientation='h',
        marker_color=shap_df['color']
    ))
    fig.update_layout(
        xaxis_title="Impact on Churn Risk (Red = Higher Risk)",
        yaxis=dict(autorange="reversed"),
        height=300,
        margin=dict(l=10, r=10, t=40, b=10)
    )
    st.plotly_chart(fig, use_container_width=True, config={'staticPlot': True})'''
with open("app.py","w") as file: 
    file.write(code)
```
